```python
import sklearn.datasets as datasets
import pandas as pd
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
from sklearn.tree import DecisionTreeClassifier
from six import StringIO
#import sklearn.externals.six.StringIO
# https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html
from IPython.display import Image
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.export_graphviz.html
from sklearn.tree import export_graphviz
# https://pypi.org/project/pydotplus/
# pip install pydotplus
# pip install graphviz
# sudo apt-get install graphviz
import pydotplus
```

```python
# متغییر مستقل
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
# متغییر وابسته
y = iris.target
y
```

```python
dtree = DecisionTreeClassifier()
# MNIST در این بخش ما 50 درصد داده ها رو میذاشتیم برای یادگیری یا فیت و 50 درصد باقیمانده رو میذاشتیم برای تست یا پیش بینی
# ولی در این بخش یا همون درخت تصمیم گیری تمام داده ها رو برای یادگیری قرار میدهیم
dtree.fit(df, y)
```

```python
pip install graphviz
```

```python
dtree.predict([
    [6.1, 2.8, 1.8, 1.9],
    [8.1, 2.1, 1.9, 2.9]
])
```

```python
dotData = StringIO()
export_graphviz(dtree, out_file=dotData, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dotData.getvalue())
Image(graph.create_png())
# https://towardsdatascience.com/gini-impurity-measure-dbd3878ead33
# https://victorzhou.com/blog/gini-impurity/#:~:text=Gini%20Impurity%20is%20the%20probability,1%E2%88%92p(i))
```
